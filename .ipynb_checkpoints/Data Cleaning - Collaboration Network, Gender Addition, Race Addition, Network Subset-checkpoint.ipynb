```python
import pandas as pd
import itertools
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
artist_data = pd.read_csv('archive/data.csv ')
```

```python
artist_master = artist_data['artists'].tolist()
release_year =  artist_data['year'].tolist()
populairty_list = artist_data['popularity'].tolist()
```

```python
artist_master
```

```python
''.join(list(artist_master[1].replace('[','').replace(']',''))).split(',')
```

```python
artist_link_df = pd.DataFrame()
source_list = []
destination_list = []
release_date = []
popularity_list_local = []

for index in range(len(artist_master)):
    entry = artist_master[index]
    entry_modified = ''.join(list(entry.replace('[','').replace(']','')))
    if ',' in entry_modified:
        entries_collab = entry_modified.split(',')
        for i, j in itertools.combinations(entries_collab, 2):
            source_list.append(i.lstrip())
            destination_list.append(j.lstrip())
            release_date.append(release_year[index])
            popularity_list_local.append(populairty_list[index])
            
artist_link_df['node_1'] = source_list
artist_link_df['node_2'] = destination_list
artist_link_df['release_date'] = release_date
artist_link_df['populairty'] = popularity_list_local
```

```python
len(artist_link_df)
```

```python
len(set(source_list).union(set(destination_list)))
```

```python
artist_link_df_unique = artist_link_df.drop_duplicates(subset=['node_1','node_2'])
```

```python
artist_link_df_unique.to_csv('archive/artist_link_df_unique.csv')
```

# Gender 

#### GitHub: https://pypi.org/project/gender-guesser/

```python
# !pip install gender_guesser
```

```python
import gender_guesser.detector as gender

d = gender.Detector()
```

```python
artist_w_genre = pd.read_csv('archive/Collaboration_nodes_w_genre.csv') # Have all 12000
```

```python
gender_predicted_list = []

for index, row in artist_w_genre.iterrows():
    prediction_successful = False
    predicted_gender_list = []
    for entries in row['Id'].split(' '):
        predicted_gender = d.get_gender(entries)
        if predicted_gender != 'unknown':
            if 'female' in predicted_gender:
                predicted_gender_list.append('female')
            elif 'male' in predicted_gender:
                predicted_gender_list.append('male')
            elif 'andy' in predicted_gender:
                predicted_gender_list.append('androgynous')
            prediction_successful = True

    if prediction_successful == True:
        gender_predicted_list.append(predicted_gender_list[0])
    else:
        gender_predicted_list.append('unknown')
```

```python
artist_w_genre['gender'] = gender_predicted_list
```

```python
artist_w_genre.to_csv('archive/Collaboration_nodes_w_genre.csv')
```

```python
print('male, female, andy and unknown counts: ', gender_predicted_list.count('male'), gender_predicted_list.count('female'), gender_predicted_list.count('androgynous'), gender_predicted_list.count('unknown'))
```

```python
plt.figure(figsize=(10,5))
plt.bar(['Male','Unknown','Female','Androgynous'], [2100,1897,996,149])
plt.xticks(rotation='vertical')
plt.ylabel('Artist count')
plt.xlabel('Gender')
plt.show()
```

```python
# !pip install ethnicolr
```

```python
from ethnicolr import census_ln, pred_census_ln
```

#### api is Asian and Pacific Islander

```python
artist_w_genre_gender = pd.read_csv('archive/Collaboration_nodes_w_genre.csv')
```

```python
artist_w_genre_gender_race = pred_census_ln(artist_w_genre_gender, 'Id')
```

```python
artist_w_genre_gender_race
```

```python
artist_w_genre_gender_race.to_csv('archive/Collaboration_nodes_w_genre.csv')
```

```python
print('Different races: ', set(artist_w_genre_gender_race['race'].tolist()))
```

```python
list_race = artist_w_genre_gender_race['race'].tolist()
distinct_race = set(list_race)
race_artist_count_dict = dict()

for race in distinct_race:
    race_artist_count_dict[race] = list_race.count(race)
```

```python
race_artist_count_dict_sorted = dict(sorted(race_artist_count_dict.items(), reverse=True, key=lambda item: item[1]))
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(race_artist_count_dict_sorted.keys()), list(race_artist_count_dict_sorted.values()))
plt.xticks(rotation='vertical')
plt.ylabel('Artist count')
plt.xlabel('Race')
plt.show()
```
