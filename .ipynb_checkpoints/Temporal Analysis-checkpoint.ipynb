```python
import pandas as pd
import itertools
from tqdm import tqdm
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
link_table = pd.read_csv('archive/Collaboration_links_w_genre.csv')
```

```python
dinstinct_artists = list(set(link_table['source'].tolist()).union(set(link_table['target'].tolist())))
```

```python
artist_year_dict = dict()

for artist in tqdm(dinstinct_artists):
    years_list = []
    artist_in_source = link_table[link_table['source']==artist] 
    artist_in_target = link_table[link_table['target']==artist]
    for year in artist_in_source['release_date'].tolist():
        years_list.append(year)
    for year in artist_in_target['release_date'].tolist():
        years_list.append(year)
    artist_year_dict[artist] = years_list
```

```python
artist_year_dict
```

```python
artist_active_year_count = dict()

for artist in artist_year_dict.keys():
    artist_active_year_count[artist] = len(artist_year_dict[artist])
```

```python
artist_active_year_count_sorted = dict(sorted(artist_active_year_count.items(), reverse=True, key=lambda item: item[1]))
```

```python
plt.figure(figsize=(20,10))
plt.bar(list(artist_active_year_count_sorted.keys())[0:20], list(artist_active_year_count_sorted.values())[0:20])
plt.xticks(rotation='vertical')
plt.ylabel('Artist count')
plt.xlabel('Race')
plt.show()
```

```python
artist_active_distinct_year_count = dict()

for artist in artist_year_dict.keys():
    artist_active_distinct_year_count[artist] = len(set(artist_year_dict[artist]))
```

```python
artist_active_distinct_year_count_sorted = dict(sorted(artist_active_distinct_year_count.items(), reverse=True, key=lambda item: item[1]))
```

```python
plt.figure(figsize=(20,10))
plt.bar(list(artist_active_distinct_year_count_sorted.keys())[0:20], list(artist_active_distinct_year_count_sorted.values())[0:20])
plt.xticks(rotation='vertical')
plt.ylabel('Artist count')
plt.xlabel('Race')
plt.show()
```

```python
def temporal_collab(artist_year_dict,link_table,artist):
    print('Worked between the years: ', min(artist_year_dict[artist]), max(artist_year_dict[artist]))
    distinct_years = set(artist_year_dict[artist])
    print('Total entries of years: ', len(artist_year_dict[artist]))
    print('Distinct entries of years: ', len(artist_year_dict[artist]))
    
    link_table_artist = link_table[link_table['source'].isin([artist])]
    link_table_artist = link_table_artist[link_table_artist['source'].isin([artist])]

    collab_number_year = []

    for year in distinct_years:
        collab_number_year.append(len(link_table_artist[link_table_artist['release_date']==year]))
        
    plt.figure(figsize=(20,10))
    plt.bar(list(distinct_years),collab_number_year)
    plt.xticks(rotation='vertical')
    plt.ylabel('Collaboration count')
    plt.xlabel('Year')
    plt.show()
    
    return 
```

```python
temporal_collab(artist_year_dict,link_table,'JAY-Z')
```

```python
temporal_collab(artist_year_dict,link_table,'Snoop Dogg')
```
